```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
df = pd.read_csv('basic_info.csv')
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
is_spanish = []
for rows in df['Nationality']:
    if rows == 'Spain':
        is_spanish.append(1)
    else:
        is_spanish.append(0)
len(is_spanish)
```

```python
df['is_spanish'] = is_spanish
df
```

```python
df = df.drop(['Contract', 'Name', 'Nationality', 'Club'], axis=1)
df
```

```python
def clean_data_money(df):
    temp_data = df.str.replace('[M, K, €]', '').astype(float)
    for values, indx in zip(df, df.index):
        if values[-1] == 'M':
            temp_data[indx] = temp_data[indx] * 1000000
        elif values[-1] == 'K':
            temp_data[indx] = temp_data[indx] * 1000
        else:
            pass
    return temp_data
df['Value'] = clean_data_money(df['Value'])
df['Wage'] = clean_data_money(df['Wage'])
df
```

```python
X = df.drop('is_spanish', axis=1)
y = df['is_spanish']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt

stats = X.shape[1]
plt.barh(range(stats), model.feature_importances_)
plt.yticks(np.arange(stats), X.columns)
plt.xlabel('Importancia de las stats')
plt.ylabel('stats')
plt.show()
```
