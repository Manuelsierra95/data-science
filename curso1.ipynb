```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('basic_info.csv')
```

```python
## Hacer grafica de jugadores des españa o fuera como (0,1)
## y replicar la grafica de arriba separada
# Age vs Overall con legend Spain(0,1)
# Age vs Potential con legend Spain(0,1)
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df['Nationality'].head(10)
```

```python
is_spanish = []
for i in range(len(df)):
    if(df.iloc[i]['Nationality'] == 'Spain'):
        is_spanish.append(1)
    else:
        is_spanish.append(0)
        
```

```python
df['is_spanish'] = np.array(is_spanish)
df
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)
# Config ax1
scatter = ax1.scatter(x=df['Age'],
                     y=df['Overall'],
                     c=df['is_spanish'],
                     cmap='summer')
ax1.set(title='Age vs Overall',
       xlabel='Age',
       ylabel='Overall')
ax1.legend(*scatter.legend_elements(),
            title='From Spain')
ax1.axhline(df['Overall'].mean(),
           linestyle='--')

# Config ax2
scatter = ax2.scatter(x=df['Age'],
                     y=df['Potential'],
                     c=df['is_spanish'],
                     cmap='winter')
ax2.set(title='Age vs Potential',
       xlabel='Age',
       ylabel='Potential')
ax2.legend(*scatter.legend_elements(), title='From Spain')
ax2.axhline(df['Potential'].mean(),
           linestyle='--')
```
